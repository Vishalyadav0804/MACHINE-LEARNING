```python
x=np.arange(1,100)
y=x**2
plt.title('Squares')
plt.ylabel('x**2')
plt.xlabel('x-values')
plt.grid() #divides the graph into square boxes called grid.
plt.xlim(-5,15)
plt.ylim(-10,100)
plt.plot(x,y, label='quadratic')
plt.plot(x,x+50,'r--', label='linear', linewidth=4)
plt.legend() #for showing quadratic and linear text in graph
plt.text(7.5,50,'Point')
plt.show()
```

```python
#sigmoid(x)=1/1+e^-x
#can be generated by math.exp(x)
```

```python
sigmoid=lambda x: 1/(1+math.exp(-x))
x=np.arange(-100,100)
y=[sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```
