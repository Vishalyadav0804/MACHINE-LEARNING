```python
import numpy as np
import matplotlib.pyplot as plt
import cv2  #for performing operation on image
%matplotlib inline
plt.rcParams['figure.figsize']=[15,10]#size reshape
```

```python
im=cv2.imread('iStock-545347988.jpg')
```

```python
(h,w,c)=im.shape
(h,w,c)
```

```python
image=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack([im,image]))  #to print both images
plt.show()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
flatImg=image.reshape(h*w,3)
clt=MiniBatchKMeans(n_clusters=16)
lables=clt.fit_predict(flatImg)
```

```python
lables.shape
```

```python
clt.cluster_centers_.astype('uint8')
```

```python
quant=clt.cluster_centers_.astype('uint8')[lables]
quant.shape
```

```python
quant=quant.reshape((h,w,3))
plt.imshow(quant)
plt.show()
```

```python
quant=cv2.cvtColor(quant,cv2.COLOR_RGB2BGR)
cv2.imwrite('iStock-545347988.jpg',quant)
```
