# PAIR-GRID

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist) # maps only diagonal elements
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',palette='husl')
x=x.map_diag(plt.hist) 
x=x.map_offdiag(plt.scatter)
x=x.add_legend() # it gives marking that pink colour is for setosa and all
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist,histtype='step',linewidth=2) 
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species') #only 2 variables 
x=x.map(plt.scatter) 
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'],hue='species') # different variables
x=x.map(plt.scatter) 
x=x.add_legend()
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
```

```python
x=sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter,s=30,edgecolor='black')
x=x.add_legend()
```

# CLUSTERMAP

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights=flights.pivot('month','year','passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False) # cluster only rows
```

```python
sns.clustermap(flights, row_cluster=False) # cluster only columns
```

```python
sns.clustermap(flights, cmap='coolwarm',linewidth=1) # cmap is for colour change
```

```python
sns.clustermap(flights, cmap='Blues_r',linewidth=1.5) 
```

```python
sns.clustermap(flights,figsize=(6,4)) # figsize is for resizing the map
```

```python
# standardize across col or rows
```

```python
sns.clustermap(flights,standard_scale=1)
```

```python
sns.clustermap(flights,standard_scale=0)
```

```python
# normalising our dataset = z_score = 0(rows)/1(columns)
```

```python
sns.clustermap(flights,z_score=0)
```
