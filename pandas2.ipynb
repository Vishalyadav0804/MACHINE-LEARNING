# Restructuring of data

```python
weather=pd.read_csv('data/weather.csv')
weather.head()
```

```python
#multiple columns are melt and stored into one
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],
       var_name='day',
       value_name='temp')
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(index=['id','year','month','day'],
                        columns='element',
                        values='temp',
                        dropna=False).reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('output/weather_tidy.csv')

#it will save this table in output file
```

```python
weather_tidy.loc[(weather_tidy.month==1) & (weather_tidy.day=='d15'),['tmax','tmin']]
```

```python
def mean_func(arr):
    return sum(arr)/len(arr)
```

```python
weather_tidy[['month','year']].apply(mean_func)
#function is applied cloumn wise
```

```python
weather_tidy[['month','year']].apply(mean_func,axis=1).head()
#function is applied row wise
```

```python
def extract(x):
    d=x[0]
    return int(d[1:])
```

```python
weather_tidy[['day']].apply(extract,axis=1)
```

```python
days=weather_tidy[['day']].apply(extract,axis=1)
days.head()
```

```python
weather_tidy['day']=days
weather_tidy.head()
```
