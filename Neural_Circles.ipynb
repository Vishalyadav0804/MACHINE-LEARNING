```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=50,factor=0.4) #factor determine radius
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
#using logistic
```

```python
from sklearn.linear_model import LogisticRegression
ls=LogisticRegression()
ls.fit(X_train,y_train)
ls.score(X_test,y_test)*100
```

```python
# using KNN
```

```python
X,y=make_circles(n_samples=50,factor=0.4) #as factor changes KNN % changes
plt.scatter(X[:,0],X[:,1],c=y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)*100
```

```python
#using  Neural-network
```

```python
X,y=make_circles(n_samples=50,factor=0.4)
plt.scatter(X[:,0],X[:,1],c=y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(solver='lbfgs',
                 alpha=1e-5,
                 hidden_layer_sizes=(6),
                 random_state=100,
                 activation='logistic',
                 max_iter=200)
mlp.fit(X_train,y_train)
mlp.score(X_test,y_test)*100
```

```python
X,y=make_circles(n_samples=500,factor=0.9)
plt.scatter(X[:,0],X[:,1],c=y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(solver='lbfgs',  #optimiser-->gradient descent--schostic gradient descend--adam--lbfs
                 alpha=1e-5,  #10^-5
                 hidden_layer_sizes=(6),
                 random_state=100,
                 activation='logistic',
                 max_iter=200)
mlp.fit(X_train,y_train)
mlp.score(X_test,y_test)*100
```

```python
#LOSS/ERROR FUNCTION
MSE-mean squared error
BCE-binary cross entropy--for two classes
CCE-catagorical cross entropy--for two or more than two classes
#Activation function
functions that introduce non-linearity eg.
1) Sigmoid/logistic function  
2)ReLU(Rectified Linear Unit)=max(0,x)--if x -ve output is 0 and if x is +ve then output is x  
3)tanh  
4)identity--> y=x
5)softmax
```
